# Matplotlib基础使用

```python
import matplotlib.pyplot as plt
import numpy as np
```

### 折线图

```python
x = np.linspace(0,10,100)
print(x)
```

```python
y = np.sin(x)
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
cosy = np.cos(x)
siny = y.copy()
plt.plot(x,cosy,color="green",label="cosx")
plt.plot(x,siny,color="red",label="siny")
plt.xlabel("x Value")
plt.legend()
plt.show()
```

### 散点图

```python
plt.scatter(x,siny)
plt.show()

x = np.random.normal(0,1,3000)
y = np.random.normal(0,1,3000) 

plt.scatter(x,y,alpha=0.5)
plt.show()
```
