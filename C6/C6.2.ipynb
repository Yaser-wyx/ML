# 梯度下降模拟

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-6,7,150)
plot_y = (plot_x-0.5)**2-3
plt.plot(plot_x,plot_y)
plt.show()
```

```python
def dJ(theta):
    return 2*(theta-0.5)

def J(theta):
    return (theta-0.5)**2-3

eta = 0.1
epsilon = 1e-10
theta = 0.0
while True:
    gradient = dJ(theta) # 求导
    last_theta = theta
    theta = theta - eta * gradient # 新的theta
    if abs( J(theta) - J(last_theta) ) <= epsilon:
        break
```

```python
plt.plot(plot_x,plot_y)
plt.scatter(theta,J(theta),color='r')
plt.show()
```

```python
print(theta)
print(J(theta))
```

```python
eta = 0.1
epsilon = 1e-10
theta = -3.0
theta_history=[theta]
while True:
    gradient = dJ(theta) # 求导
    last_theta = theta
    theta = theta - eta * gradient # 新的theta
    theta_history.append(theta)
    if abs( J(theta) - J(last_theta) ) <= epsilon:
        break
```

```python
plt.plot(plot_x,plot_y)
plt.plot(np.array(theta_history),J(np.array(theta_history)),color='r',marker='+')
plt.show()
```
