# 随机梯度下降法

```python
import numpy as np
import matplotlib.pyplot as plt
m = 10000
x = np.random.normal(size=m)
X = np.array(x.reshape(-1,1))
y = x * 3. + 4. + np.random.normal(0 , 3,size=m)
print(y.shape)
plt.scatter(X,y)
plt.show()
```

```python
# 求单个特征的梯度
def dJ(X_b_i,y_i,theta):
    return X_b_i.T.dot(X_b_i.dot(theta)-y_i) * 2

def gsd(X_b,y,n_iters,inital_theta):
    theta = inital_theta
    t0,t1=5,50
    def learning_rate(t):
        return t0/(t1+t)
    
    for i in range(n_iters):
        random_i = np.random.randint(len(X_b)) # 选取随机的一行
        gradient = dJ(X_b[random_i],y[random_i],theta) # 求取该行的梯度值
        theta = theta - learning_rate(i) * gradient # 根据这随机的样本，进行梯度计算
    
    return theta
```

```python
X_b = np.hstack([np.ones(len(X)).reshape(-1,1),X])
inital_theta = np.zeros(X_b.shape[1])
theta = gsd(X_b,y,m*3,inital_theta)
print(theta)
```

```python
plt.scatter(X,y)
y_p = X_b.dot(theta)
plt.plot(x,y_p,color='r')
plt.show()
```
