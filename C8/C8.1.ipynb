# 8.1多项式回归

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt 
```

```python
np.random.seed(666)
x = np.random.uniform(-3,3,500)
y = 0.5 * x**2 +3.5 * x + 1.1 +np.random.normal(0,1, 500)
plt.scatter(x,y)
plt.show()
```

```python
temp = x.reshape(-1,1)
X = np.hstack([temp,temp**2])
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
y_predict = lin_reg.predict(X)
plt.scatter(x,y)
plt.plot(np.sort(x),y_predict[np.argsort(x)],color='r')
plt.show()
```

```python
print(lin_reg.coef_)
print(lin_reg.intercept_)
```
